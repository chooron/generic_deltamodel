```python
# unpack pickle file

import pickle
import numpy as np
import pandas as pd
import torch

# Load the pickle file
with open("/data/yxs275/CAMELS/training_file", "rb") as f:
    data = pickle.load(f)
    
# Check the type and get the shape
if isinstance(data, np.ndarray):
    print("Shape:", data.shape)
elif isinstance(data, pd.DataFrame):
    print("Shape:", data.shape)  # (rows, columns)
elif isinstance(data, torch.Tensor):
    print("Shape:", data.shape)
elif isinstance(data, list):  
    print("Length of list:", len(data))
    if len(data) > 0 and hasattr(data[0], "shape"):  
        print("First element shape:", data[0].shape)  # Handle list of arrays/tensors
else:
    print("Type not recognized, check manually:", type(data))

```
